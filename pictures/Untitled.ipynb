```python
%matplotlib inline
import numpy
import pylab
```

# Time difference value

```python
w_bin = 1
fps = 30
period = numpy.floor(1000./fps)

h = 10
t_vals = numpy.arange(33)
h_vals = (period - t_vals)
fig_vals = pylab.figure()
pylab.plot(t_vals, h_vals)
pylab.show()
```

```python
t_last = 10

t_now = numpy.arange(11, 100)

tdiff = t_now - (t_last - (period - t_last)*w_bin)
```

```python
h_num = period - numpy.mod(tdiff, period)
fig_hnum = pylab.figure()
pylab.plot(t_now, h_num)
pylab.show()

    
```
